```python
import pandas as pd 

# padas :  used for exploring and manuipilating data

melb_data = pd.read_csv("../../csv/melb_data.csv")

melb_data.describe()
```

```python
# list out the columns in dataframe

melb_data.columns

```

```python
# Drop missing value column or row wise uidng dropna

melb_data
```

```python
melb_data = melb_data.dropna(axis=0)
```

```python
y = melb_data["Price"]
```

```python
y
```

```python
input_features =  ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

X = melb_data[input_features]

X
```

```python
from sklearn.tree import DecisionTreeRegressor

# Deefine model

model = DecisionTreeRegressor(random_state=1)

# fit the model

model.fit(X,y)
```

```python
preds = model.predict(X)
preds
```

```python
# Evaluate the model
# metrics for summarizing model qualiity
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y,preds)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(X,y,random_state=1)

new_model = DecisionTreeRegressor(random_state=1)
new_model.fit(X_train,y_train)

preds = new_model.predict(X_valid)
```

```python
mean_absolute_error(y_valid,preds)
```

```python
from sklearn.ensemble import RandomForestRegressor
def get_mea(max_leaf_nodes, X_train,X_valid,y_train,y_valid):
    model = RandomForestRegressor(random_state=1, max_leaf_nodes=max_leaf_nodes)
    model.fit(X_train,y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid,preds)
    
```

```python
for i in [5, 50, 500,1000,3000, 5000]:
    print(get_mea(i, X_train,X_valid,y_train,y_valid))
```
