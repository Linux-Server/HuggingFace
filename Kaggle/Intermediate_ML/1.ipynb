```python
import pandas as pd 
```

```python
# Three approaches dealing with missing value
# drop the colums
# Imputation
# extension with Imputation
# impuation will applcable to integers

from sklearn.model_selection import train_test_split

melb_data = pd.read_csv("../../csv/melb_data.csv")
melb_data.head()
```

```python
y = melb_data["Price"]
y
```

```python
# Now drop Price from dataFrame

melb_data =melb_data.drop("Price", axis=1)
```

```python
melb_data
```

```python
X = melb_data.select_dtypes(exclude=["object"])
X
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=1)
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
# Approach 1 : Drop missing columns
# get the name of columns with mising value

cols_missing = [columns for columns in X_train.columns if X_train[columns].isnull().any()]

cols_missing

reduced_X_train = X_train.drop(cols_missing, axis=1)
reduced_X_valid = X_valid.drop(cols_missing, axis=1)
```

```python
X_train["BuildingArea"].head().isnull()
```

```python
X_train["BuildingArea"].head().isnull().all()
```

```python
print("MAE from Approach 1 (Drop columns with missing values):")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

```python
X_train
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()

imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
```

```python
imputed_X_train
```

```python
print("MAE from Approach 2 (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

```python
# Score from Approach 3 (An Extension to Imputation)¶
# Make copy to avoid changing original data (when imputing)
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()
```

```python
X_train_plus
```

```python
# find columns with missing value
cols_missing = [col for col in X_train if X_train[col].isnull().any()]
cols_missing
```

```python
# Make new columns indicating what will be imputed
X_train_plus
```

```python
for col in X_train_plus[cols_missing]:
    X_train_plus[col + "_was_missing"] = X_train_plus[col].isnull()
    X_valid_plus[col + "_was_missing"] = X_valid_plus[col].isnull()
    
X_train_plus
```

```python
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.transform(X_valid_plus))

imputed_X_train_plus
```

```python
# Imputation removed column names; put them back
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns

print("MAE from Approach 3 (An Extension to Imputation):")
print(score_dataset(imputed_X_train_plus, imputed_X_valid_plus, y_train, y_valid))
```
